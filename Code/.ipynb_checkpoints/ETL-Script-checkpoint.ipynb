```python
# Call on dependencies

import pandas as pd
import requests
import json
from pprint import pprint
from config import api_key, connection_string
from config import connection_string

```

```python
# Import & read CSV file

data1 = '../Netflix_Data/User1.csv'

data2 = '../Netflix_Data/User2.csv'

data3 = '../Netflix_Data/User3.csv'

data4 = '../Netflix_Data/User4.csv'

# Create dataframe based on CSV file

user1 = pd.read_csv(data1)

user2 = pd.read_csv(data2)

user3 = pd.read_csv(data3)

user4 = pd.read_csv(data4)

```

```python
#Creating lists of movie / show titles for API calls

user1_list = user1["Title"].to_list()

user2_list = user2["Title"].to_list()

user3_list = user3["Title"].to_list()

user4_list = user4["Title"].to_list()
```

```python
#Adding lists together

movie_list = user1_list + user2_list + user3_list + user4_list
len(movie_list)
```

```python
#Converting to dictionary back to list to remove duplicated

movie_list = list(dict.fromkeys(movie_list))

len(movie_list)
```

```python
#API calls

url = "http://www.omdbapi.com/?apikey=" + api_key + "&t="

responses = [];

for movie in movie_list:
    movie_data = requests.get(url + movie).json()
    responses.append(movie_data)
```

```python
#Verifying movies were recognized 

responses[5]
```

```python
#Checking JSON format for further manipulation

responses
```

```python
#Checking keys to determine which fields to pull for Postgres load

responses[1].keys()
```

```python
#Looping through JSON output, creating lists to load to DataFrame
#Incorporating Try-Except to filter out titles without a match

titles = []
years = []
rateds = []
releaseds = []
genres = []
types = []

for mov in responses:
    try:
        titles.append(mov['Title'])
        years.append(mov['Year'])
        rateds.append(mov['Rated'])
        releaseds.append(mov['Released'])
        genres.append(mov['Genre'])
        types.append(mov['Type'])
    except KeyError:
        pass

```

```python
#Validating output

titles
```

```python
#Creating dataframe from gathered lists

df = pd.DataFrame(list(zip(titles,rateds,genres,types)),columns=['title','rated','genre','type'])
```

```python
#Determining if shows made the list

df.type.value_counts()
```

```python
#Making the dataframe movies only

df = df[df.type != 'series']
```

```python
#Verifying the DF is movies only

df.type.value_counts()
```

```python
#Manipulating for Postgres import

df = df.set_index("title")
```

```python
#Dropping movie distinction as all rows are movies

df = df.drop(columns=['type'])
```

```python
#Verifying no duplicates in DataFrame

df = df.drop_duplicates()
```

```python
#Checking DataFrame

df.head()
```

```python
#Importing dependencies for Postgres upload

import psycopg2
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
#Creating engine

engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Loading DF to 'moviedata' in Postgres
#Results show error because load already occurred

df.to_sql(name='moviedata', con=engine, if_exists='append', index=True)
```
